```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

# Activation Function
## Step function
$$
h(x) = 
\begin{cases}
1 \quad (x > 0) \\
0 \quad (x <= 0)
\end{cases}
$$

## Sigmoid function
$$
h(x) = \frac{1}{1 + e^{-x}}
$$

## ReLU (Rectified Linear Unit)
$$
h(x) = 
\begin{cases}
x \quad (x > 0) \\
0 \quad (x <= 0)
\end{cases}
$$

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_ylim(-0.1, 1.1)
ax.set_title("step function")
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_ylim(-0.1, 1.1)
ax.set_title("sigmoid function")
```

```python
x = np.arange(-5.0, 5.0, 0.1)
sigmoid_y = sigmoid(x)
step_y = step_function(x)
fig, ax = plt.subplots()
ax.plot(x, sigmoid_y, label="sigmoid", linestyle='-')
ax.plot(x, step_y, label="step", linestyle=':')
ax.set_ylim(-0.1, 1.1)
ax.legend()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_ylim(-0.1, 5.5)
ax.set_title("ReLU")
```
